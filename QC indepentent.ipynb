# Model for Axis bank

```python
import ollama

sentiment = []
for sentence in var.astype(str):
    prompt = ("""
            You are the most advanced AI model in the world. I will provide a comment. Your task is to analyze it and categorize the sentiment using the rules below.
            Follow these rules step by step, in order, for every comment.
            
            
            CRITICAL EXCLUSION (APPLIES FIRST):
                If the comment is related to stock price, share market, Sensex, mutual funds, NSE Nifty, or financial performance (including technical analysis, trading, or price changes), ALWAYS return SENTIMENT: ONE-PIECE, regardless of any other content.
                Examples:
                "Axis Bank share price falls 5% today." → SENTIMENT: ONE-PIECE
                "NSE Nifty 50: Top 5 stock losers... Axis Bank..." → SENTIMENT: ONE-PIECE
                "Axis Bank stock analysis for June 05, 2025." → SENTIMENT: ONE-PIECE
            
            Categorization Rules (Step-by-Step):
                1. SENTIMENT: CODE-RED
                Return SENTIMENT: CODE-RED if any of the following conditions are met:
                Axis Bank is directly referenced in connection with a crime, illegal activity, fraud, scam, or criminal investigation.
                Any major mishappening (such as fire, death, major accident, or other serious incidents) occurs inside Axis Bank premises or is directly related to Axis Bank operations.
                Do NOT return CODE-RED if the comment is about stock price, share market, Sensex, mutual funds, or financial performance.
                If the crime or incident only occurred near an Axis Bank branch or office but is not related to Axis Bank itself, do NOT return CODE-RED; proceed to step 3.
                
                2. SENTIMENT: AXIS BANK
                Return SENTIMENT: AXIS BANK if ALL of the following are true:
                The comment is about Axis Bank or directly mentions the names of its higher management (current or former) or Axis Bank launches a new product (card or schemes).
                The comment is important enough to be brought to their attention, but does not involve crime or customer support.
                The comment announces or discusses a partnership, collaboration, or joint initiative involving Axis Bank (including with other companies, government, NGOs, or institutions), except if it is about a crime, mishappening, or financial performance.
                Any mention of Axis Bank in the context of CSR (corporate social responsibility) activities.
                The comment is NOT related to stock price, share market, Sensex, mutual funds, NSE Nifty, or financial market commentary.
                If the comment is promotional, marketing, or offer-related content, do NOT return AXIS BANK; proceed to step 3.
                If the comment is about Axis Bank subsidiaries or services other than Axis Bank, do NOT return AXIS BANK; proceed to step 3.
            
                3. SENTIMENT: ONE-PIECE
                Return SENTIMENT: ONE-PIECE if the comment falls into any of the categories below:
                Customer support issues (for Axis Bank or any Axis Group entity).
                Mentions of any subsidiary or service under Axis Ltd other than Axis Bank, including:

                    Axis Max Life Insurance

                    Axis Capital

                    Axis Securities

                    Axis Finance

                    Axis Mutual Fund

                    Axis Trustee Services

                    Axis Direct

                    Axis Private Equity

                    Axis AMC / Axis Asset Management Company

                    Axis Pension Fund

                Any other Axis Group company or service (excluding Axis Bank).
                Crimes or mishappenings that occurred near an Axis Bank branch, but not involving Axis Bank.
                Promotional, marketing, or offer-related content (treat as unrelated).
                Personal experiences related to Axis Bank (even if negative or emotional).
                Any unrelated or general topic, including scientific, technical, or medical content (even if the word "axis" appears but is not about Axis Bank).
                Any comment related to stock price, share market, Sensex, mutual funds, NSE Nifty, or financial performance.
              
            Response Format
            Only respond with the sentiment label using this format:
            SENTIMENT: {CODE-RED / AXIS BANK / ONE-PIECE}
            Do not include explanations,answer any or any additional text.
            Default Rule
            If a comment does not clearly fit the rules for CODE-RED or AXIS BANK, ALWAYS default to ONE-PIECE.

          """    
         + sentence
    )
    response = ollama.generate(model='mistral:latest', prompt=prompt)
    print(sentence,'::',response['response'])
    sentiment.append(response['response'])
data['Related_to_axisbank'] = sentiment
```

```python
data['Related_to_axisbank'].value_counts()
```

```python
data['Related_to_axisbank'] = data['Related_to_axisbank'].str.strip()
axis_bank = data[data['Related_to_axisbank'] == 'SENTIMENT: AXIS BANK']
code_red = data[data['Related_to_axisbank'] == 'SENTIMENT: CODE-RED']
with pd.ExcelWriter('/Users/abhishekchauhan/Downloads/downloads/Axis_bank_28.xlsx',engine='openpyxl') as writer:
    axis_bank.to_excel(writer, index=False,sheet_name='AxisBank')
    code_red.to_excel(writer, index=False,sheet_name='CodeRed')
```
